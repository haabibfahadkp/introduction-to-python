```python
x=[10,20,30,50,60,70,80,90,11]
k=0
for i in x:
  k=k+i
  print(k)
```

```python
x=[10,20,30,40,50,60,70,80,90]
k=0
for i in x:
  if i!=30:
    k=k+i
    print(k)
```
